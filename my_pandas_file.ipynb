- When using row names or column names, use 'loc'
- When using index values on the basis of elements, use 'iloc'
- When using iloc, it doesn't matter whatever can be the names of rows & columns are, we can still use numbers for referencing them.

```python
newdf.iloc[[0,5],[1,2]]    # (0,5) = rows & (1,2) = columns
```
