```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_json('flipkart_fashion_products_dataset.json')
df.head()
```

```python
df.info()
```

```python
df['actual_price'].str.replace(',','')  #remove comma
```

```python
def clean_price(value):
    if not value:
        return np.nan
    if isinstance(value,str):
        value = value.replace(',','')
        if value.isnumeric():
           return float(value)
    return value    
    # print(value)
df['actual_price'].apply(clean_price)  #test
```

```python
df['actual_price'] = df['actual_price'].apply(clean_price) #actual implementation
df['selling_price'] = df['selling_price'].apply(clean_price)  #actual implementation
df.info()
```

```python
def clean_discount(value):
    if not value:
        return np.nan
    elif isinstance(value,str):
        value = value.replace('% off','')
        try:
            value = float(value)
            return value
        except:
            print(value)  # get an idea of what value failed
            return np.nan
    else:
        return value
    
#test
df['discount'].apply(clean_discount)    
```

```python
df['discount'] = df['discount'].apply(clean_discount) #actual implementation
```

```python
def clean_rating(value):
    if value:
        return float(value)
    return np.nan

#test
df['average_rating'].apply(clean_rating)
```

```python
df['average_rating'] = df['average_rating'].apply(clean_rating)  #actual implementation
```
